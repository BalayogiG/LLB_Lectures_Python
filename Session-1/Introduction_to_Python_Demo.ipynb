<b>List operations</b>

```python
# Declaring List
# List can contain any number of values.
# List can have different kinds of data types.
# list counting start with 0
myList = [1,2.5,'hello',True,1]
print(myList)

# To print specific value
print(myList[1])
```

<b>Tuple Operations</b>

```python
# Declaring tuple
# Tuple can contain duplicate values
# Tuple is immutable (you can't change the values in it)
# The values only can be accessed with index value
myTuple= (10,20,30,10,40,50,50)
print('Complete tuple: ',myTuple)

# To count the values in tuples
count_of_10 = myTuple.count(10)
print('Numeber of 10 in tuple: ',count_of_10)

# To get the index of value in tuple
index_of_20 = myTuple.index(20)
print('Index of value 20:',index_of_20)
```

<b>Set operations</b>

```python
# Declaring set
# Set doesn't have indices
# Set shouldn't contain any duplicate values.
mySet = {10,20,30,40,50}
print('Complete set: ',mySet)

# To check the set contains the specific value
# This will print either true or false
print('Is 10 contains in set?: ',10 in mySet)
```

```python
# To add a value to Set
mySet.add('80')
print('Updated set: ',mySet)

# To add multiple values to Set
mySet.update([90,60,100,130,150])
print('Updated set with more values: ',mySet)
```

```python
# To remove a value from Set
mySet.remove('80')
print('Set without 80: ',mySet)
```

```python
set1 = {10, 20, 30}
set2 = {10,30,50}

# This will return false
print('Is set1 is a subset of set2: ',set1.issubset(set2))

# This will return a set with the union of the two sets.
print('Union of two sets ',set1.union(set2))
```

```python
geneset1 = {'PTEN','KRAS','BRCA1','ULK1','DEC1'}
geneset2 = {'DEC1','KRAS','ABL1'}
print('Union of geneset1 and geneset2: ',geneset1.union(geneset2))
print('Intersection of geneset1 and geneset2: ',geneset1.intersection(geneset2)) #Intersection
```

<b>Loops</b>

<font size='3'>
    <p> To perform repetitive computation for a finite number of time.</p>
<ul>
    <li>while loop</li>
    <li>for loop</li>
    <li>nested loop</li>
</ul>
</font>

```python
# while loop
count = 0
while count < 9:
    print('The count is: ',count)
    count  = count + 1 # this can also be write by count += 1
print('good bye')
```

```python
# for loop
gene1 = 'atagcatca'

for i in gene1:
    print('current letter',i)    
```

```python
# nested loop
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
    for y in fruits:
        print(x, y)
```

<b>Control keywords</b>

<font size='3'>
    <p> To change the flow of the program execution.</p>
<ul>
    <li>continue</li>
    <li>break</li>
    <li>pass</li>
</ul>
</font>

```python
# continue keyword
# Use to go to the first line of the code
gene = 'ATAGACTA'
for i in gene:
    if i == 'T':
        continue
    print('current letter: ',i)
```

```python
# break keyword
# Use to stop the program execution
gene = 'ATAGACTA'
for i in gene:
    if i == 'T':
        break
    print('current letter: ',i)
```

```python
# pass keyword
# Use to pass even the condition is true
gene = 'ATAGACTA'
for i in gene:
    if i == 'T':
        pass
    print('current letter: ',i)
```
