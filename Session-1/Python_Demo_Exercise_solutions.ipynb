<b>Question 1</b>

Write a Python program to count the number of characters (character frequency) in a string. <br>
<b>Sample String :</b> google.com'<br>
<b>Expected Result :</b> {'o': 3, 'g': 2, '.': 1, 'e': 1, 'l': 1, 'm': 1, 'c': 1}<br>

```python
# Complete the code and run

def char_frequency(str1):
    dict = {}
    for n in str1:
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
        else:
            dict[n] = 1
    return dict
print(char_frequency('google.com'))
```

<b>Question 2</b

Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '#', except the first char itself.<br>
<b>Sample String :</b> 'restart'<br>
<b>Expected Result :</b> 'resta#t'<br>

```python
def change_char(str1):
    char = str1[0]
    str1 = str1.replace(char, '#')
    str1 = char + str1[1:]

    return str1

print(change_char('restart'))
```

<b>Question 3</b>

Write a Python function that takes a list of words and returns the length of the longest one.

```python
# complete the code and run
def find_longest_word(words_list):
    word_len = []
    for n in words_list:
        word_len.append((len(n), n))
    word_len.sort()
    return word_len[-1][1]

print(find_longest_word(["ATA", "ATAGCA", "AGAT"]))
```
