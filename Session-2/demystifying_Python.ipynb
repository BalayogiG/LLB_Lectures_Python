<b>File handling Modes</b>

```python
# File handling
# File creation and writing

f = open('hello.txt','w+')
for i in range(5):
    f.write("Bioinformatics\n")
f.close()
```

```python
t = open("good.txt",'w+b')
byte_arr = [120,3,255,0,100]
binary_format = bytearray(byte_arr)
t.write(binary_format)
t.close()
```

```python
# Empty file creation

f = open('demofile.txt','x')
f.close()
```

```python
# file reading and printing

f = open('hello.txt','r')
contents = f.read()
print(contents)
```

```python
# file appending and printing

f = open('hello.txt','a+')
for i in range(2):
    f.write("Appended text: Genes and enzymes\n")
f.close()

f = open('hello.txt','r')
contents = f.read()
print(contents)
```
